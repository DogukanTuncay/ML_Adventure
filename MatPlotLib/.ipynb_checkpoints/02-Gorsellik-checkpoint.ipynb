```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyArray1 = np.linspace(0,10,20)
```

```python
numpyArray2 = numpyArray1 ** 2
```

```python
(benimFigur,benimEksen) = plt.subplots()
benimEksen.plot(numpyArray1,numpyArray2,color="#000",alpha=0.3)
benimEksen.plot(numpyArray2,numpyArray1,color="#C96F23")
```

```python
(yeniFigur,yeniEksen) = plt.subplots()
yeniEksen.plot(numpyArray1,numpyArray1 +2,color="blue",linewidth=1.0)
yeniEksen.plot(numpyArray1,numpyArray1 +3,color="yellow", linewidth = 0.5)
yeniEksen.plot(numpyArray1,numpyArray1 +4,color="yellow", linewidth = 4,linestyle=":")
yeniEksen.plot(numpyArray1,numpyArray1 +5,color="red", linewidth = 4,linestyle="--")
yeniEksen.plot(numpyArray1,numpyArray1 +6,color="green", linewidth = 4,linestyle="-.")
yeniEksen.plot(numpyArray1,numpyArray1 +7,color="#000000", linestyle="--",marker="o",markerfacecolor="red",markersize=8)
yeniEksen.plot(numpyArray1,numpyArray1 +8,color="#000000", linestyle="--",marker="+",markersize=4)
```

## scatter

```python
plt.scatter(numpyArray1,numpyArray2)
```
