```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
# bunu yazarak plt.show() kullanımına gerek olmadna grafiklerin gösterilmesini sağlayabiliriz.
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,74,70,90,95,100]
```

```python
numpyYasListesi = np.array(yasListesi)
```

```python
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyKiloListesi,numpyYasListesi,"b")
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"b")
plt.xlabel("Yaş Verisi")
plt.ylabel("Kilo Verisi")
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 **3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
```

```python
plt.subplot(1,2,1) # 1. satır 1. kolon 1. grafik
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1. satır 2. kolon 2. grafik
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.1,0.1,0.3,0.3])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni veri ismi")
figureAxes.set_ylabel("Y ekseni veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9]) #birinci parametre x ekseninde 0-1 aralığındaki yeri, ikinci parametre y ekseninde aynı. Üç ve dört ise boyut width ve height
eksen2 = figure2.add_axes([0.3,0.5,0.4,0.4])
eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"r")
eksen2.set_xlabel("X ekseni Küçük")
eksen2.set_ylabel("Y ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(myFigure,myEksenler) = plt.subplots(nrows =1,ncols = 2)
type(myEksenler)
```

```python
yeniFigur = plt.figure(dpi=100)
yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi2,label="NumpyDizisi ** 3")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label="NumpyDizisi ** 2")
yeniEksen.legend(loc=1)
```

```python
yeniFigur.savefig("benimFigur.png",dpi=500) # dpi çözünürlük, yüksek daha kaliteli. Fazla yüksek olunca yapımı ve çıktı vermesi çok uzun sürebilir.
```
